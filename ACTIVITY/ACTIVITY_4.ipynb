**Gaussian Pyramid with Edge Detection**

This program demonstrates multi-scale image processing using the Gaussian Pyramid technique.

**1: Level 0 – Original Image**

Displays the full-resolution input image.

Fine details (textures, small objects) are clearly visible.

Edge detection captures maximum detail.

**2: Level 1 – Half Size**

Image size is reduced to 50%.

Some fine details begin to disappear.

Edges become slightly smoother.

**3: Level 2 – Quarter Size**

Resolution decreases further.

Small textures and tiny edges fade away.

Only major object boundaries remain clear.

**4: Level 3 – One-Eighth Size**

Very low resolution image.

Only large structural features are visible.

Edge detection highlights strong, large-scale edges only.

```python
!pip install opencv-python pillow
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
level0 = img
level1 = cv2.pyrDown(level0)
level2 = cv2.pyrDown(level1)
level3 = cv2.pyrDown(level2)
def detect_edges(image):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, 100, 200)
    return edges
edges0 = detect_edges(level0)
edges1 = detect_edges(level1)
edges2 = detect_edges(level2)
edges3 = detect_edges(level3)
plt.figure(figsize=(16,10))
plt.subplot(4,2,1); plt.imshow(level0); plt.title("Level 0 - Original")
plt.axis("off")
plt.subplot(4,2,2); plt.imshow(edges0, cmap="gray"); plt.title("Edges - Level 0")
plt.axis("off")
plt.subplot(4,2,3); plt.imshow(level1); plt.title("Level 1 - Half Size")
plt.axis("off")
plt.subplot(4,2,4); plt.imshow(edges1, cmap="gray"); plt.title("Edges - Level 1")
plt.axis("off")
plt.subplot(4,2,5); plt.imshow(level2); plt.title("Level 2 - Quarter Size")
plt.axis("off")
plt.subplot(4,2,6); plt.imshow(edges2, cmap="gray"); plt.title("Edges - Level 2")
plt.axis("off")
plt.subplot(4,2,7); plt.imshow(level3); plt.title("Level 3 - One-Eighth Size")
plt.axis("off")
plt.subplot(4,2,8); plt.imshow(edges3, cmap="gray"); plt.title("Edges - Level 3")
plt.axis("off")
plt.tight_layout()
plt.show()
print("✅ Gaussian Pyramid and Edge Detection Completed!")
```
